```python
# calculate the mean vaule for "math_score" column
avg_Math_Score = df["Math_Score"].astype("float").mean(axis=0)
print("Average of Math Score= ", avg_Math_Score)

df["Math_Score"] = df["Math_Score"].replace(np.nan, avg_Math_Score)
```

```python
print(df["Math_Score"].isnull().sum())
```

```python
avg_Reading_Score = df["Reading_Score"].astype("float").mean(axis=0)
print("Average of Reading Score= ", avg_Reading_Score)

df["Reading_Score"] = df["Reading_Score"].replace(np.nan, avg_Reading_Score)
```

```python
avg_Writing_Score = df["Writing_Score"].astype("float").mean(axis=0)
print("Average of Writing Score= ", avg_Writing_Score)

df["Writing_Score"] = df["Writing_Score"].replace(np.nan, avg_Writing_Score)
```

```python
df.isnull().sum()
```

```python
df.drop("Test_Preparation", axis=1, inplace=True)
```

```python
df["Math_Reading_Total"] = df["Math_Score"] + df["Reading_Score"]
```
